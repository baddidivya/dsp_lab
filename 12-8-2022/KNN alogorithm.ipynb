```python
p=np.linspace(train["asset_cost"].min(),train["asset_cost"].max(),10)
```

```python
for i in range(len(train["asset_cost"])):
    if train.loc[i,"asset_cost"]>=p[0] and train.loc[i,"asset_cost"]<=p[1]:
        train.loc[i,"asset_cost"]=1
    elif train.loc[i,"asset_cost"]>=p[1] and train.loc[i,"asset_cost"]<=p[2]:
        train.loc[i,"asset_cost"]=2
    elif train.loc[i,"asset_cost"]>=p[2] and train.loc[i,"asset_cost"]<=p[3]:
        train.loc[i,"asset_cost"]=3
    elif train.loc[i,"asset_cost"]>=p[3] and train.loc[i,"asset_cost"]<=p[4]:
        train.loc[i,"asset_cost"]=4
    elif train.loc[i,"asset_cost"]>=p[4] and train.loc[i,"asset_cost"]<=p[5]:
        train.loc[i,"asset_cost"]=5
    elif train.loc[i,"asset_cost"]>=p[5] and train.loc[i,"asset_cost"]<=p[6]:
        train.loc[i,"asset_cost"]=6
    elif train.loc[i,"asset_cost"]>=p[6] and train.loc[i,"asset_cost"]<=p[7]:
        train.loc[i,"asset_cost"]=7
    elif train.loc[i,"asset_cost"]>=p[7] and train.loc[i,"asset_cost"]<=p[8]:
        train.loc[i,"asset_cost"]=8
    else:
        train.loc[i,"asset_cost"]=9
```

```python
l=[]                
for i in range(len(train["no_of_loans"])):
    if train.loc[i,"no_of_loans"]==0 and train.loc[i,"no_of_curr_loans"]==0:
        l.append(-1)
    else :
        l.append(train.loc[i,"no_of_loans"]-train.loc[i,"no_of_curr_loans"])
train["paid"]=l
```
