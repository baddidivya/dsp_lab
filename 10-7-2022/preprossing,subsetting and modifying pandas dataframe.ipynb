# Data Farmatting

```python
dict={"Name":["Football","FB","basket ball","bb"],
      "num":[11,9,9,11]
     }
df=pd.DataFrame(dict)
df
```

```python
mapping={
    "Football":"FB",
    "FB":"FB",
    "basket ball":"bb",
    "bb":"bb"
}
```

```python
df.Name=df.Name.map(mapping)
df
```

```python
df["number"]=[110,100,120,140]
df
```

```python
df["score"]=df.number.apply(lambda x:x/10)  #create a new column by using existing column
df
```
