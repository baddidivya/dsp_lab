```python
import matplotlib.pyplot as plt
import numpy as np
plt.title('line plot')
x=np.array([1,2,3,4])
y=np.array([10,11,12,13])
name='x-axis'
sumu='y-axis'
plt.xlabel(name)
plt.ylabel(sumu)
plt.grid()
plt.plot(x,y,"--",linewidth=2,mec="r",marker="P")
plt.show()
```

```python
sub=["Dsp","cd","Coa","Wt"]
marks=[10,15,12,20]
plt.bar(sub,marks)
plt.title("Sub vs marks")
plt.xlabel("Sub")
plt.ylabel("Marks")
plt.show()
```

```python
plt.barh(sub,marks)
plt.title("Subject vs marks")
plt.xlabel("marks")
plt.ylabel("subjects")
plt.show()
```

```python
plt.hist(marks,color="r")
plt.xlabel("sub")
plt.ylabel("marks")
plt.show()
```

```python
plt.pie(marks,labels=["Dsp","cd","coa","wt"])
plt.legend(title="subjects")
plt.show()
```

```python
temp=np.array([320,400,120,500,190])
days=np.array(["day1","day2","day3","day4","day5"])
plt.subplot(2,2,1)
co=["b","hotpink","g","y","r"]
plt.pie(temp,colors=co,labels=days)
plt.legend(title="Temp")
plt.subplot(2,2,2)
plt.plot(days,temp)
plt.title("Temperature In this week")
plt.xlabel("Days")
plt.ylabel("Temp")
plt.subplot(2,2,3)
plt.bar(days,temp)
plt.title("Temperature In this week")
plt.xlabel("Days")
plt.ylabel("Temp")
plt.subplot(2,2,4)
plt.barh(days,temp)
plt.title("Temperature In this week")
plt.xlabel("Days")
plt.ylabel("Temp")
plt.show()

```
