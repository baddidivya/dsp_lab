```python
import matplotlib.pyplot as plt
import numpy as np
plt.title('line plot')
x=np.array([1,2,3,4])
y=np.array([10,11,12,13])
name='x-axis'
sumu='y-axis'
plt.xlabel(name)
plt.ylabel(sumu)
plt.plot(x,y)
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
x=np.array([1,10,3,4,5,6])
y=np.array([10,2,11,3,13,4])
plt.plot(x,y)
plt.show()
```

```python
plt.plot(y,marker="o",mec="r",color="blue")
plt.show()
```

```python
plt.plot(x,linewidth=10,marker="*",mec="g",color="b")
plt.show()
```
