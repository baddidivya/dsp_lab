```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x=np.array([5,15,25,35,45,55])
print(x)
x=np.array([5,15,25,35,45,55]).reshape(-1,1)
y=np.array([5,20,14,32,22,38])
print(y)
plt.scatter(x,y)
```

```python
model=LinearRegression()
model.fit(x,y)
print(model.intercept_)
print(model.coef_)
print(model.score(x,y))
```

```python
y_pre=model.predict(x)
print(y_pre)
```

```python
print(model.intercept_+model.coef_*x)
```

```python
print(model.score(x,y_pre))
plt.scatter(x,y)
plt.plot(x,y_pre)
```

```python
x_new=np.arange(5).reshape(-1,1)
print(x_new)
```

```python
y_new=model.predict(x_new)
print(y_new)
```

```python
plt.scatter(x_new,y_new)
plt.plot(x_new,y_new)
print(model.score(x_new,y_new))
```

```python
x1=np.array([1,2,3,4,5])
y1=np.array([4,5,6,7,8])
print(x1)
x1=x1[:,np.newaxis]
print(x1)
print(y1)
model=LinearRegression().fit(x1,y1)
```

```python
y_pre1=model.predict(x1)
print(y_pre1)
```

```python
plt.scatter(x1,y1)
plt.plot(x1,y_pre1)
print(model.score(x1,y_pre1))
```

```python
data=pd.read_csv("samplex.csv")
data
```

```python
plt.scatter(data["sales"],data["price"])
plt.xlabel("sales")
plt.ylabel("price")
plt.show()
```

```python
a=data["sales"].values.reshape(-1,1)
b=data["price"].values
reg=LinearRegression().fit(a,b)
print(reg.score(a,b))
print(reg.coef_)
print(reg.intercept_)
plt.scatter(a,b)
plt.plot(a,b)
```

```python
b_pre=reg.predict(a)
print(reg.score(a,b_pre))
plt.scatter(a,b)
plt.plot(a,b_pre)
```
